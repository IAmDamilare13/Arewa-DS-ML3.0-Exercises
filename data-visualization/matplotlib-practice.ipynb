```python
# Figure Size

# Wide width and short height (good for time series)
plt.figure(figsize=(10, 3)) 
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Wide Plot")
plt.show()

# Perfect square plot (good for scatter plot)
plt.figure(figsize=(6, 6)) 
plt.scatter([1, 2, 3], [4, 5, 6])
plt.title("Square Plot")
plt.show()

# Tall and narrow plot (Good for stacked subplots)
plt.figure(figsize=(5, 10))
# Add three vertical subplots
for i in range(3):
    plt.subplot(3, 1, i+1)
    plt.plot([1, 2, 3], [i, i*2, i*3])
plt.tight_layout()
plt.show()
```

```python
# Create a figure with 3 subplots (arranged vertically)

fig = plt.figure(figsize=(4, 4)) # 8" wide, 12" tall

# Line Plot
ax1 = fig.add_subplot(1, 1, 1) # 3 rows, 1 column, position 1

#data
random_walk = np.random.standard_normal(50).cumsum()

ax1.plot(random_walk, color="black", linestyle="dashed")
```

```python
# Histogram (Middle)

fig = plt.figure(figsize=(8, 12)) # 8" wide, 12" tall
ax2 = fig.add_subplot(3, 1, 2) # Position 2
normal_data = np.random.standard_normal(100)
ax2.hist(normal_data, bins=20, color="black", alpha=0.3)
ax2.set_title("Distribution Histogram")
ax2.set_ylabel("Frequency")
```

```python
# Scatter PLot
fig = plt.figure(figsize=(8, 12)) # 8" wide, 12" tall

ax3 = fig.add_subplot(3, 1, 3) # Position 3
x_values = np.arange(30)
y_values = x_values + 3 * np.random.standard_normal(30)
ax3.scatter(x_values, y_values)
ax3.set_title("Noisy Linear Relationship")
ax3.set_xlabel("X Values")
ax3.set_ylabel("Y Values")

plt.tight_layout() # Prevent label overlapping
plt.show()
```

```python
# Three Figures

fig = plt.figure(figsize=(8, 12)) # 8" wide 12" tall

ax1 = fig.add_subplot(3, 1, 1) # 3 rows, 1 column, position 1
ax1.plot([1, 2, 3, 4], [1, 4, 9, 16], 'r-') # Red line
ax1.set_title("Quadratic Growth")
ax1.set_ylabel("Y Values")

ax2 = fig.add_subplot(3, 1, 2) # 3 rows, 1 column, position 2
ax2.bar(['A', 'B', 'C'], [3, 7, 2], color='blue') # Blue bars
ax2.set_title("Category Comparison")
ax2.set_ylabel("Counts")

ax3 = fig.add_subplot(3, 1, 3) # 3 rows, 1 column, position 3
ax3.scatter(np.random.rand(20), np.random.rand(20), color='green') # Green dots
ax3.set_title("Random Scatter Plot")
ax3.set_xlabel("X Values")
ax3.set_ylabel("Y Values")

# Final adjustments
plt.tight_layout() # Automatically space subplots nicely
plt.show()
```

```python
# Colour, Markers and Line Styles

x = np.linspace(0, 10, 100)
y = np.sin(x)

# Different color specifications
plt.figure(figsize=(10, 5))
plt.plot(x, y, color='b') # Named color
plt.plot(x, np.sin(x-1), color='g') # Short color code (g=green)
plt.plot(x, np.sin(x-2), color='0.75') # Grayscalde (0=black, 1=white)
plt.plot(x, np.sin(x-3), color='#FF0000') # Hex code (red)
plt.plot(x, np.sin(x-4), color=(1,0,1)) # RGB tuple (purple)
plt.plot(x, np.sin(x-5), color='chartreuse') # CSS color name
plt.title("Different Color Specifications")
plt.show()
```

```python
# Line Styles

# Create sample data
x = np.linspace(0, 10, 100)

# Create figure and plot
plt.figure(figsize=(10, 5))

# Plot with different line styles
plt.plot(x, np.sin(x), linestyle='-', marker='o', label='solid')
plt.plot(x, np.sin(x-1), linestyle='--', label='dashed')
plt.plot(x, np.sin(x-2), linestyle=':', label='dotted')
plt.plot(x, np.sin(x-3), linestyle='-.', label='dashdot')

# Add title and legend
plt.title("Different Line Styles")
plt.legend()
```

```python
# Markers

plt.figure(figsize=(10, 5))
plt.plot(x[:15], np.sin(x[:15]), marker='o') # Circle
```

```python
# Different Marker Styles
plt.plot(x[:15], np.sin(x[:15]-1), marker='v') # Triangle down
plt.plot(x[:15], np.sin(x[:15]-2), marker='^') # Triangle up
plt.plot(x[:15], np.sin(x[:15]-3), marker='s') # Square
plt.plot(x[:15], np.sin(x[:15]-4), marker='*') # Star
plt.plot(x[:15], np.sin(x[:15]-5), marker='+') # Plus
plt.plot(x[:15], np.sin(x[:15]-6), marker='.') # Point


plt.title("Different Marker Styles")
plt.legend(['o: Circle', 'v: Triangle down', '^: Triangle up', 
            's: Square', '*: Star', '+: Pluse', '.: Point'])
plt.show()
```

```python
# Combinning All Three - color line style and markers

plt.figure(figsize=(10, 5))

plt.plot(x, np.sin(x),
         color='red',
         marker='o',
         linestyle='-')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x, np.sin(x), 'r-o')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x, np.sin(x), 'r-o')
plt.plot(x, np.cos(x), 'g--s') # Green dashed line with square markers
plt.plot(x, -np.sin(x), 'b:.') # Blue dotted line with point markers
plt.title('Combining Color Line Style, and Markers')
plt.legend(['r-o: red solid circles', 'g--s: green dashed squares',
            'b:.: blue dotted points'])
```

```python
# Control the scale and appearance of axis markers

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)

# Customize ticks
plt.xticks([0, 2, 4, 6, 8, 10]) # Set x-axis tick position
plt.yticks([-1, -0.5, 0, 0.5, 1],
           ['Low', '', 'Mid', '', 'High']) # Custom labels
plt.show()
```

```python
# Add decriptive test to axes

plt.plot(x, y)
plt.xlabel("Time (seconds)", fontsize=12, color='blue', rotation=0) # Xaxis label
plt.ylabel("Voltage (mV)", rotation=90) # Rotated y-label
plt.title("Signal Measurement", pad=10) # Title with padding

plt.show()
```

```python
# Legends

# Create sample data
x = np.linspace(0, 10, 100)

# Create figure and plot
plt.figure(figsize=(10, 5))

# Plot with different line styles
plt.plot(x, np.sin(x), linestyle='solid', label='solid') # or '-'
plt.plot(x, np.sin(x-1), linestyle='dashed', label='dashed') # or '--'
plt.plot(x, np.sin(x-2), linestyle='dotted', label='dotted') # or ':'
plt.plot(x, np.sin(x-3), linestyle='dashdot', label='dashdot') # or '-.'

# Add title and legend
plt.title("Different Line Styles")
plt.legend(loc='best') # Add title to legend
```
