In this assignment, dig deeper into the dataset to discover a fact about a given type of bird. Visualize all the interesting data you can uncover about Snow Geese. Use the three plots mentioned above to tell a story in your notebook.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import dataset
birds = pd.read_csv('../data/birds.csv')

birds.head()
```

## Line Chart

### Conversation Status of Birds in Minnesota

```python
# Filter birds by conservation status
conv_stat = birds['ConservationStatus'].value_counts()

# Plot line chart
conv_stat.plot(kind='line', marker='o')
plt.title("Plot of Birds by Conservation Status")
plt.ylabel("Number of Birds")
```

Among the six conservation status of birds in Minnesota, larger number of birds are least concern, including the Snow Geese bird.

        CR: Critically Endangered
        EN: Endangered
        EX: Extinct
        LC: Least Concern
        NT: Near Threatened
        VU: Vulnerable

## Scatter Plot

### Wingspan by Genus

```python
# Filter birds in Snow Geese family - Anatidae
sg_family = birds[birds.Family.isin(['Anatidae'])]

# Show a scatter plot show genues with maximum wing span
sns.relplot(x='MaxWingspan', y='Genus', data = sg_family)
plt.title("Plot of Genues by Maximum Wingspan")
plt.style.use("ggplot")
```

The plot above shows that birds in the Snow Goose genus i.e. Anser are of medium sizes as minimum wingspan ranges from 120 to 170 (SI unit unknown).

```python
# Show a scatter plot show genues with minimum wingspan
sns.relplot(x='MinWingspan', y='Genus', data = sg_family)
plt.title("Plot of Genues by Minimum Wingspan")
plt.style.use("ggplot")
```

The plot above shows that birds in the Snow Goose genus i.e. Anser are of medium sizes as maximum wingspan ranges from 120 to 140 (SI unit unknown).

## Bar Plot



```python
# Filter birds by category
cat_min_length = birds.groupby('Category')['MaxLength'].mean().sort_values(ascending=False)

# Plot bar chart of birds by category
cat_min_length.head(15).plot(kind='barh')
plt.title("Plot of Birds by Category")
plt.style.use("ggplot")
```
