# Euro 2012 Data

### Filtering and Sorting Data

```python
# Import the necessary libraries
import numpy as np
import pandas as pd

# Import the dataset
data = '../data/Euro_2012_stats_TEAM.csv'

# Assign it to a variable called euro12
euro12 = pd.read_csv(data)

# Select only the goal column
goals_scored = euro12['Goals']

# How many teams participated in the euro2012
total_teams = len(euro12['Team'])
print(f'There are only {total_teams} team who participated.')

# What is the number of columns in the dataset
total_columns = len(euro12.columns)
print(f'\nThere is a total of {total_columns} columns in the dataset.')

# View only the columns Team, Yellow Cards and Red Cards and assign them to a datafram called discipline
discipline = euro12[['Team', 'Yellow Cards','Red Cards']]

# Sort the teams by Red Cards, then to Yellow Cards
discipline_sorted = discipline.sort_values(['Red Cards', 'Yellow Cards'], ascending=False)
discipline_sorted
```

```python
# Calculate the mean Yellow Cards given per Team
mean_yellow_card = discipline_sorted[['Yellow Cards']].mean()[0]
print(f'The mean yellow card given per team is: {mean_yellow_card:.2f}.')

# euro12.columns

# Filter teams that scored more than 6 goals
more_than_6_goals = euro12[euro12['Goals'] > 6] # filter teams with above 6 goals
more_than_6_goals = more_than_6_goals[['Team', 'Goals']] # Subset team and goals

# Select the teams that start with G
teams_starting_with_g = euro12[euro12['Team'].str.startswith('G')]
print(f'The teams starting with "G" are:')
teams_starting_with_g.iloc[:,0]
```

```python
# Select the first 7 columns
first_seven_columns = euro12.iloc[:,0:7]
print(f'\nThe first seven columns are:')
first_seven_columns
```

```python
# Select all column except the last 3
all_columns_except_last_3 = euro12.iloc[:,0:-3]
print(f'\nThe first seven columns are:')
all_columns_except_last_3
```
